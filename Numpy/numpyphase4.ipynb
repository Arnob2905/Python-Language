```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
```

```python
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.random.rand(3,3)
arr3=np.zeros((4,4))
```

```python
#Save files in new text
np.save("arr1.npy",arr1)
np.save("arr2.npy",arr2)
np.save("arr3.npy",arr3)
```

```python
#loaded
loaded_arr1=np.load("arr1.npy")
print(loaded_arr1)
loaded_arr2=np.load("arr2.npy")
print(loaded_arr2)
loaded_arr3=np.load("arr3.npy")
print(loaded_arr3)
```

```python
#loading
try:
    # logo = np.load("numpy_logo.npy")  this use for loading..

    logo = mpimg.imread("numpy_logo.npy.png")  # or "numpy_logo.npy.png" if really named that

    # Display
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("NumPy Logo")
    plt.grid(False)  # Grid off + no axes
    
    #for displaying dark logo..
    dark_logo = 1 - logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("NumPy Dark Logo")
    plt.grid(False)

except FileNotFoundError:
    print("NumPy logo file not found.")
```
