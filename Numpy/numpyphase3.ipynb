```python
# Data structure :  [restaurant_id, 2021, 2022 , 2023, 2024]
sales_data= np.array([
    [1, 150000, 180000, 220000, 250000], #Paradise Biryani
    [2, 120000, 140000, 160000, 190000], #Beijing Bites
    [3, 200000, 230000, 260000, 300000], #Pizza Hub
    [4, 180000, 210000, 240000, 270000], #Burger Point
    [5, 160000, 185000, 205000, 230000]  #Chai Point
])

print("=== Zomato sales analysis ===")
print("\n Sales Data Shape ", sales_data.shape)
print("\n Sample Data for 1st 3 restaurant: ", sales_data[0:3])
```

```python
#Total Sales per year
print(np.sum(sales_data, axis=0))

yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)
```

```python
#Minimum sales per restaurant..
min_sales= np.min(sales_data[:, 1:], axis=1)
print(min_sales)

#Maximum sales per year..
max_sales=np.max(sales_data[:, 1:],axis=0)  #top to bottom  rows
print(max_sales)

#Average sales per restaurant..
avg_sales=np.mean(sales_data[:, 1:], axis=1)   #Left to right cols
print(avg_sales)
```

```python
#Cumulative Sales..

cumsum=np.cumsum(sales_data[:, 1:],axis=1)
print(cumsum)

#Matplotlib..
plt.figure(figsize=(6,5))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales per year ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print(vector1 + vector2)      #Addition
print(vector1 - vector2)      #Substraction
print(vector1 * vector2)      #Multiplication
print(vector2 / vector1)      #Division
print(np.dot(vector1,vector2)) #Dot product
angle=np.arccos(np.dot(vector1,vector2)/ (np.linalg.norm(vector1) * np.linalg.norm(vector2)))  #angle
print(angle)
```

```python
#Vectorized Operations..
rest_types=np.array(["biryani","chinese","pizza","burger","cafe"])
vectorized_upper = np.vectorize(str.upper)
print("Vectorized Upper",vectorized_upper(rest_types))
```

```python
#Broadcasting

monthly_avg = sales_data[:, 1:] / 12

print(monthly_avg)
plt.figure(figsize=(6,5))
plt.plot(monthly_avg)
plt.title("Monthly Average")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

```
