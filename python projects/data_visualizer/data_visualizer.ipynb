# Importing Libraries.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
```

# Initialization

```python
class SalesDataAnalyzer:
    def __init__(self):
        self.data = None

    def load_data(self):
        file_path = input(" Enter the CSV file path or name: ")
        try:
            self.data = pd.read_csv(file_path)
            print(" Dataset loaded successfully!")
        except FileNotFoundError:
            print("File not found! Please check the path and try again.")

    def explore_data(self):
        print(" First 5 Rows:")
        print(self.data.head())
        print("\n Info:")
        print(self.data.info())
        print("\nData Types:")
        print(self.data.dtypes)

    def clean_data(self):
        print("\nBefore Cleaning:")
        print(self.data.isnull().sum())
        self.data.fillna(self.data.mean(numeric_only=True), inplace=True)
        print("Missing values filled (numerical columns only).")

    def mathematical_operations(self):
        print("\nSummary Statistics:")
        print(self.data.describe())

    def visualize_data(self):
        print("\nAvailable Columns:")
        print(list(self.data.columns))

        kind = input("\nChoose chart type (Bar / Line / Scatter / Pie / Hist / Area / Heatmap / Boxplot / Violinplot): ").strip().lower()
        x = input("Enter X column ( leave blank for heatmap): ").strip()
        y = input("Enter Y column ( leave blank for pie chart or heatmap): ").strip()

        if kind not in kind:
            print("Invalid chart type.")
            return

        if kind not in["heatmap","pie"] and (x not in self.data.columns or y not in self.data.columns):
            print("Invalid column names. Try Again.")
            return

        try:
            plt.figure(figsize=(6, 5))
            title = f"{kind.title()} Chart - {x} vs {y if y else ''}"

            if kind == "bar":
                self.data.groupby(x)[y].sum().plot(kind="bar")
            elif kind == "line":
                self.data.plot(x=x, y=y, kind="line")
            elif kind == "scatter":
                self.data.plot(x=x, y=y, kind="scatter")
            elif kind == "pie":
                self.data[x].value_counts().plot.pie(autopct='%1.1f%%')
            elif kind == "hist":
                self.data[y].plot(kind="hist")
            elif kind == "area":
                self.data[[x, y]].plot.area(stacked=True)
            elif kind == "heatmap":
                corr = self.data.corr(numeric_only=True)
                sns.heatmap(corr,annot=True, cmap="magma", fmt=".2f")
                title = "Heatmap - Correlation Matrix"
            elif kind == "boxplot":
                sns.boxplot(data=self.data, x=x ,y=y)
            elif kind == "violinplot":
                sns.violinplot(data=self.data, x=x, y=y, hue=" ")            
            else:
                print("Invalid chart type.")
                return
            plt.title(title)
            plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"Error generating plot: {e}")

    def save_plot(self):
        filename = input("Enter filename to save plot (with .png): ")
        try:
            plt.savefig(filename)
            print(f"Plot saved as {filename}")
        except Exception as e:
            print(f"Could not save plot: {e}")
```

```python
analyzer = SalesDataAnalyzer()

while True:
    print("\n === DATA ANALYSIS & VISUALIZATION PROGRAM ===")
    print("1. Load Dataset")
    print("2. Explore Dataset")
    print("3. Handle Missing Data")
    print("4. Generate Summary Statistics")
    print("5. Data Visualization")
    print("6. Save Visualization")
    print("7. Exit")

    choice = input("Enter your choice (1-7): ")

    if choice == "1":
        analyzer.load_data()
    elif choice == "2":
        if analyzer.data is not None:
            analyzer.explore_data()
        else:
            print("Load data first!!!")
    elif choice == "3":
        if analyzer.data is not None:
            analyzer.clean_data()
        else:
            print("Load data first!!!")
    elif choice == "4":
        if analyzer.data is not None:
            analyzer.mathematical_operations()
        else:
            print("Load data first!!!")
    elif choice == "5":
        if analyzer.data is not None:
            analyzer.visualize_data()
        else:
            print("Load data first!!!")
    elif choice == "6":
        if analyzer.data is not None:
            analyzer.save_plot()
        else:
            print("Load data first!!!")
    elif choice == "7":
        print("Exiting...")
        break
    else:
        print("Invalid choice.")
```
